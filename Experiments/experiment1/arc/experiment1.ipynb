```python
input_folder ='/kaggle/input/dop123/test/images'
confidence_threshold = 0.5

image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff'}
output_folder='/kaggle/working/output_images'
# --- 2. Main Loop to Process Each Image ---
os.makedirs(output_folder, exist_ok=True)
# List all files in the input folder
try:
    file_list = os.listdir(input_folder)
except FileNotFoundError:
    print(f"Error: Input folder not found at '{input_folder}'")
    print("Please ensure the path to your input folder is correct.")
    exit()

print(f"Starting processing for {len(file_list)} files in '{input_folder}'...")

for filename in file_list:
    # Check if the file is an image by its extension
    file_ext = os.path.splitext(filename)[1].lower()
    if file_ext not in image_extensions:
        print(f"Skipping non-image file: {filename}")
        continue

    # Construct the full path for the input image
    image_path = os.path.join(input_folder, filename)

    try:
        # Read the image using OpenCV
        image = cv.imread(image_path)
        if image is None:
            print(f"Warning: Could not read image '{filename}'. Skipping.")
            continue
    except Exception as e:
        print(f"Error reading {filename}: {e}. Skipping.")
        continue

    # --- 3. Run Prediction ---
    results = model.predict(source=image, conf=confidence_threshold)
    result = results[0]  # Get the first result object

    # --- 4. Draw Bounding Boxes on the Image ---
    for box in result.boxes:
        coords = box.xyxy[0].tolist()
        x1, y1, x2, y2 = map(int, coords)
        confidence = box.conf[0].item()
        class_id = int(box.cls[0].item())
        class_name = model.names[class_id]
        
        color = (0, 255, 0) # Green
        
        cv.rectangle(image, (x1, y1), (x2, y2), color, 2)

    output_filename = f"{os.path.splitext(filename)[0]}_detected.jpg"
    output_path = os.path.join(output_folder, output_filename)
    
    # Save the image with the drawn boxes
    cv.imwrite(output_path, image)
```
